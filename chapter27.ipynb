```python
import pandas as pd
```

```python
df = pd.read_csv('attendance.csv')
df['Home'] = df['Home'].astype('category')
df['Away'] = df['Away'].astype('category')
```

### Question 1: Group by a categorical column and take the mean of the numeric columns

```python
df.groupby('Home').mean()
```

### Question 2: Group by a categorical column and take the mean and max of the numeric columns

```python
df.groupby('Home').agg(['mean', 'max'])
```

### Question 3: Group by a categorical column and apply a custom aggregation function that calculates the mode of the numeric columns

```python
def series_mode(ser):
    return ser.mode().iloc[0]
```

```python
df.groupby('Home').agg(series_mode)
```

### Question 4: Group by two categorical columns and take the mean of the numeric columns

```python
df.groupby(['Home', 'Away'], observed=True).agg('mean')
```

### Question 5: Group by binned numeric column and take the mean of the numeric columns.

```python
# Set up the bins to a specific size
bins = [1000, 10000, 20000, 30000, 40000, 50000, 60000]

# create a new column for the bin
df['binned'] = pd.cut(df['Attendance'], bins)

# group and average
df.groupby('binned').mean()
```
