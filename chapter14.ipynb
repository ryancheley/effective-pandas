```python
import pandas as pd
```

```python
df = pd.read_csv('attendance.csv')
```

### Question 1: Create a histogram from a numeric column. Change the bin size.

```python
df.plot.hist(bins=50)
```

### Question 2: Create a boxplot from a numeric column.

```python
min_team_total_attendance = df.groupby('Home').agg({'Attendance': 'sum'}).sort_values(by='Attendance').iloc[0, :].name
max_team_total_attendance = df.groupby('Home').agg({'Attendance': 'sum'}).sort_values(by='Attendance').iloc[-1, :].name
attendance_filter = (df['Home'] == min_team_total_attendance) | (df['Home'] == max_team_total_attendance)
df[attendance_filter].plot.box(by='Home')
```

### Question 3: Create a Kernel Density Estimate plot from a numeric column.

```python
df.plot.kde()
```

### Question 4: Create a line from a numeric column.

```python
df[attendance_filter].plot.line(by='Home')
```
