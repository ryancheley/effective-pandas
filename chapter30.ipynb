```python
import pandas as pd
from random import normalvariate
```

```python
df = pd.read_csv('attendance.csv', index_col=['Date']).sort_index()
```

### Question 1: Melt two numeric columns values into a single column. Add a new column to indicate what the values mean.

```python
df['A Random Number'] = 42
df_melt = df.melt(id_vars=['Home', 'Away'], var_name='Type', value_name='Number', value_vars=['Attendance', 'A Random Number'])
df_melt
```

### Question 2: Un-melt the above.

```python
df_melt.pivot_table(index=['Home', 'Away'], columns='Type', values='Number').reset_index()
```

### Question 3: Group by two columns, take the mean and unstack the result.

```python
df = pd.read_csv('attendance.csv', index_col=['Date']).sort_index()
df.groupby(['Home', 'Away']).mean().unstack(0)
```

### Question 4: Group by two columns, take the mean, and unstack the result, and flatten the columns.

```python
def flatten_cols(df):
    cols = ['_'.join(map(str, vals))
            for vals in df.columns.to_flat_index()]
    df.columns = cols
    return df
```

```python
df.groupby(['Home', 'Away']).mean().unstack().pipe(flatten_cols)
```
