```python
import pandas as pd
```

```python
df = pd.read_csv('attendance.csv')
```

### Question 1: Add a new column that is the sum of a numeric column that was grouped by a string column.

```python
df = df.assign(total_home_attendance=df.groupby('Home').Attendance.transform('sum'))
df
```

### Question 2: Filter out the rows that have less than 3 entries when grouped by a string column.

```python
df[df.total_home_attendance >= 2_800_000]
```
