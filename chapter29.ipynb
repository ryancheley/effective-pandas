```python
import pandas as pd
```

```python
df = pd.read_csv('attendance.csv')
```

### Question 1: Summarize the count of one categorical column against another.

```python
pd.crosstab(df['Home'], df['Away'])
```

### Question 2: Summarize the count of one categorical column against another, adding margins.

```python
pd.crosstab(df['Home'], df['Away'], margins=True)
```

### Question 3: Summarize the count of one categorical column against another in a heatmap.

```python
# teams = sorted(list(df.Home.unique()))
teams = ['Arizona Diamondbacks', 'Colorado Rockies', 'Los Angeles Dodgers', 'San Diego Padres', 'San Francisco Giants', 'Houston Astros']
(pd.crosstab(index=[df.Home],columns=[df.Away], values=df.Attendance, aggfunc='mean', margins=True)\
    .loc[
        teams, 
        teams])\
    .style.background_gradient(cmap='viridis', axis=None)
```
