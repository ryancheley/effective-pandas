# Chapter 34: Styling Dataframes

```python
import pandas as pd
```

```python
df = pd.read_csv('attendance.csv')
```

### Question 1: Color the background of the first two columns blue

```python
df.head(10).\
    style.\
        set_properties(
            **{'background-color': 'blue'}, 
            subset=pd.IndexSlice[:, ['Date', 'Away']]
            )
```

### Question 2: Format the numeric values by specifying precision and thousands separator

```python
df.head(10).\
    style.\
    format(
        {'Attendance': '{: ,}'}
        )
```

### Question 3: Include a bar plot in a column

```python
df.head().\
    style.\
    bar(
        subset=['Attendance'], 
        color='green', 
        vmax=df.Attendance.quantile(0.95), 
        align='mid'
    ).\
    format({'Attendance': '{: ,}'})
```

### Question 4: Set a background gradient for a column

```python
df.head(10).\
    style.\
        background_gradient(cmap='coolwarm', subset=['Attendance', ])
```

### Question 5: Make the column headers sticky

```python
df.head(100).\
    style.\
        background_gradient(cmap='coolwarm', subset='Attendance').\
        set_sticky(axis='columns')
```
