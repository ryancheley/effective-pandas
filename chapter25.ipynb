### Question 5: Create a bar plot from a frequency count of a categorical column

```python
df['Attendance']\
    .div(10000)\
    .round(0)\
    .astype("category")\
    .value_counts()\
    .plot\
    .bar()
```

### Question 6: Create a pie plot from a frequency count of a categorical column

```python
df['Attendance']\
    .div(10000)\
    .round(0)\
    .astype("category")\
    .value_counts()\
    .plot\
    .pie()
```
